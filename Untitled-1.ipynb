```python
#serie temporal
time_series = df['consumo']
plt.figure(figsize=(10,6))
sns.lineplot(time_series)
plt.title('distribuiçao da serie temporal')
plt.show()
```

```python
#estudo media movel
def m_avg(time_series, dias, nome_coluna):
    plt.figure(figsize=(10,6))
    m_avg = time_series.rolling(dias).mean()
    sns.lineplot(m_avg, color='red', label='media movel')
    sns.lineplot(time_series, color='blue', label='serie real')
    plt.title(f'Média movel da timeseries {nome_coluna}')
    plt.legend(loc = 'best')
    plt.show()
    
```

```python
#calculo de media movel em uma janela de 365 dias
m_avg(time_series, 365, nome_coluna='consumo')
```

```python
#sazonalidade 
def sazonalidade(time_series):
    plt.figure(figsize=(10,6))
    saz = time_series.diff()
    sns.lineplot(saz)
    plt.title('Sazonalidade da timeseries')
    plt.legend(loc='best')
    plt.show()
sazonalidade(time_series)
```

```python
#calculo de tendencia
def trend(time_series, window, min_periods):
    plt.figure(figsize=(10,6))
    trend = time_series.rolling(window=window,
                                center=True,
                                min_periods=min_periods).mean()
    
    sns.lineplot(time_series, color='blue', label='Consumo')
    sns.lineplot(trend, color='red', label='Tendencia')
    plt.title('Tendencia consumo energia eletrica Brasil')
    plt.show()

trend(time_series, 365, 183) #calculando tendencia em uma jandela de 365 dias
```

```python
#ciclos 
y_cicles = time_series.rolling(180, center=True).std()
plt.figure(figsize=(10,6))
sns.lineplot(y_cicles)
plt.title('Analise de volatidade do consumo')
plt.show()
```

```python
#separaçao das novas colunas numericas para analise de autocorrelaçao 
col_autocorrelation = df.select_dtypes(exclude='object')
col_autocorrelation
type(col_autocorrelation)
```

```python
#graficos de autocorrelaçao 
n_cols = len(col_autocorrelation.columns)  
fig, axes = plt.subplots(1, n_cols, figsize=(5 * n_cols, 5))  
for i, auto in enumerate(col_autocorrelation):
    ax = axes[i] if n_cols > 1 else axes
    plot_acf(col_autocorrelation[auto].values.squeeze(), lags=30, alpha=0.05, ax=ax)
    ax.set_title(f'autocorrelaçao: {auto}')
plt.tight_layout()
plt.show()
```
