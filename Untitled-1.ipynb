```python
#estudo media movel
def m_avg(time_series, dias, nome_coluna):
    plt.figure(figsize=(10,6))
    m_avg = time_series.rolling(dias).mean()
    sns.lineplot(m_avg, color='red', label='media movel')
    sns.lineplot(time_series, color='blue', label='serie real')
    plt.title(f'M√©dia movel da timeseries {nome_coluna}')
    plt.legend(loc = 'best')
    plt.show()
    
```

```python
#calculo de media movel em uma janela de 19 anos
m_avg(time_series, 6935, nome_coluna='consumo')
```

```python
#sazonalidade 
def sazonalidade(time_series):
    plt.figure(figsize=(10,6))
    saz = time_series.diff()
    sns.lineplot(saz)
    plt.title('Sazonalidade da timeseries')
    plt.legend(loc='best')
    plt.show()
sazonalidade(time_series)
```

```python
#calculo de tendencia
def trend(time_series, window, min_periods):
    plt.figure(figsize=(10,6))
    trend = time_series.rolling(window=window,
                                center=True,
                                min_periods=min_periods).mean()
    
    sns.lineplot(time_series, color='blue', label='Consumo')
    sns.lineplot(trend, color='red', label='Tendencia')
    plt.title('Tendencia consumo energia eletrica Brasil')
    plt.show()

trend(time_series, 6935, 3468) #calculando tendencia em uma jandela de 19 anos
```

```python
#variaveis para split
x = df.drop('consumo', axis=1)
y = df['consumo']
```

```python
#separa√ßao colunas num e cat para pipe
n_attribs = ['numero_consumidores', 
                    # 'consumo',
                    'consumo_medio_por_pessoa',
                    'variacao_de_consumo_por_pessoa']


cat_attribs = ['sigla_uf', 'tipo_consumo']
```

```python
#pre processamento
column_transf = ColumnTransformer([
    ('scaler_pca', Pipeline(steps=[
        ('scaler', RobustScaler()),
        ('pca', IncrementalPCA(n_components=3, batch_size=500))
    ]), n_attribs),  
    ('one_hot', OneHotEncoder(drop="if_binary", handle_unknown="ignore"), cat_attribs)  
])
```

```python
#teste pipe
pipe = Pipeline(steps=[
    ('transform', column_transf),
    ('model', LinearRegression())
])
pipe
```

```python
#separa√ßao x_test, y_test
train_size = 0.9  
split = int(len(x) * train_size)  
x_train, x_test = x[:split], x[split:]
y_train, y_test = y[:split], y[split:]
```

```python
x_train
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
#treino pipe
pipe.fit(x_train, y_train)
```

```python
#fazendo predicoes
pred = pipe.predict(x_test)
pred
```

```python
#fun√ßao que calcula metricas
def calculate_metrics(model, y_teste, pred):
    MAE = float(mean_absolute_error(y_teste, pred))
    R2 = float(r2_score(y_teste, pred))
    MSE = float(mean_squared_error(y_teste, pred))
    RMSE = float(np.sqrt(MSE))
    MAPE = float(mean_absolute_percentage_error(y_teste, pred))

    results = { 
        'modelo': model,
        'MAE': MAE,
        'R2_score': R2,
        'MSE': MSE,
        'RMSE': RMSE,
        'MAPE': MAPE
    }
    return results
```

```python
calculate_metrics(pipe, y_test, pred)
```

```python
#objeto com modelos 
models = {'linear_reg': LinearRegression(),
        # 'logistic_reg': LogisticRegression(solver='lbfgs'),
        'rforest_reg': RandomForestRegressor(),
        'gboost_reg': GradientBoostingRegressor(),
        'tree_reg': DecisionTreeRegressor(),
        'ada_reg': AdaBoostRegressor(),
}
```

```python
#ve a variabilidade dos dados
sns.lineplot(y_train, color='blue')
sns.lineplot(y_test, color='red')
```

```python
#ponteiro para treinar modelos e predi√ßao 
predictions = {}
for i, (model_name, model) in enumerate(models.items()):
    pipe.set_params(model=model).fit(x_train,y_train)
    pred_ = pipe.predict(x_test)
    metrics = calculate_metrics(model=model, y_teste=y_test, pred=pred_)

    print(metrics)
```

Cross validation 

```python
#defini√ßao de kf com kfolds
kf = KFold(n_splits=5, shuffle=True)
```

```python
#cross val score
for i, cross_ in models.items():
    model_cross_val= pipe.set_params(model=cross_)
    cross_val = cross_val_score(model_cross_val, X=x_train, y=y_train, cv=kf, scoring='r2')
    print(f'metricas de cross do modelo {cross_} e R¬≤:{cross_val}, com o valor max de: {cross_val.max()}')
```
